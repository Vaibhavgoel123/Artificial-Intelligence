```python
import torch
import matplotlib.pyplot as plt
```

```python
def plot_results(model, distances, times):
  """
  Plots the actual data points and the model's predicted line for a given dataset.
  Args:
  model: The trained machine learning model to use for predictions.
  distances: The input data points (features) for the model.
  times: The target data points (labels) for the plot.
  """
  model.eval() #Evaluation mode

  # Disabling gradient calculation for efficient interference
  with torch.no_grad():

    #Predictions using trained model
    predicted_times = model(distances)

    #Creating figure for plot
    plt.figure(figsize=(11,9))

    #Plotting Actual data...
    plt.plot(distances.numpy(), times.numpy(), color = 'blue', marker = '1', linestyle = 'None', label = 'Actual delivery times')

    #Plotting Predicted data
    plt.plot(distances.numpy(), predicted_times.numpy(), color = 'red', marker = '1', linestyle = 'None', label = 'Predicted delivery times')

    #Plot title
    plt.title("Actual vs Predicted")

    #Setting x and y lablels
    plt.xlabel("Distance in miles")
    plt.ylabel("Time in minutes")

    #Displaying the legend
    plt.legend()

    #Displaying the grid
    plt.grid(True)

    #Show the plot
    plt.show()
```
