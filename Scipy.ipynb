## Scipy Optimizers
Optimizers are the set of procedures defined in Scipy that either find the minimum value of function or root of an equation.

Find the root of ``` x + cos(x) ``` function.

### Roots of an Equation
```optimize.root()``` function is used to calcualte the root.  
It has two parameters:  
***fun***: a function representing an equation.  
***x0***: an initial guess for root.

````python
from scipy.optimize import root
from math import cos
def eqn(x):
    return x+cos(x)

myroot = root(eqn,0)
print(myroot.x)
````

````python
print(myroot)
````

### Minimizing the functions
A function represents a curve with *High points* and *Low points*.  
High points are known as ***Maxima***.  
Low points are known as ***Minima***.  
The highest point in the whole curve is known as ***Global Maxima*** whereas rest are ***Local Maxima***.  
The lowest point in the whole curve is known as ***Global Minima*** whereas rest are ***Local Minima***.  


```scipy.optimize.minimize()``` function is used to minimize the function.  
The ```minimize()``` takes following arguments:    
***fun*** - a function representing equation.  
***x0*** - an initial guess for the root.  
***method*** - name of the method to use.  
Some of the methods are:  
1. ```CG``` : ***Conjugate Gradient***
2. ```BFGS``` : ***Quasi-Newton***
3. ```Newton-CG``` : ***Newton Conjugate Gradient***
4. ```L-BFGS-B```: ***Limited Memory BFGS with Bounds***
5. ```TNC``` : ***Truncated Newton***
6. ```COBYLA``` : ***Constrained Optimization BY Linear Approximations***
7. ```SLSQP``` : ***Sequential Least Squares Programming***

Only ```COBYLA``` is not gradient-based, rest others are gradient-based. 

***callback*** - function called after each iteration of optimization.  
***options*** - a dictionary defining extra parameters:  
```python
{
    "disp" : boolean - print detailed description.
    "gtol" : number - the tolerance of the error.
}
```

````python
from scipy.optimize import minimize
def eqn(x):
    return x**2+x+2
mymin = minimize(eqn,0,method='BFGS')
print(mymin)
````

## Scipy Sparse Data
***Sparse data*** is the data that has mostly unused elements (elements that doesn't carry any information). It can be considered as dataset consists of zeroes at high rate.  
***Dense array*** is the opposite of sparse array: most of the values are not zero.  

Scipy has ```scipy.sparse``` module that provides functions to deal with sparse data.  
They are primarily of two types:  
**CSC** - Compressed Sparse Column. For efficient arithmetic, fast column slicing.  
**CSR** - Compressed Sparse Row. For fast row slicing, faster matrix vector products.

````python
import numpy as np
from scipy.sparse import csr_matrix
arr = np.array([0,0,0,0,0,1,1,0,2])
print(csr_matrix(arr))
````

````python
arr = np.array([[0,0,0],[0,0,1],[1,0,2]])
print(csr_matrix(arr).data)
````

````python
print(csr_matrix(arr).count_nonzero())
````

````python
mat = csr_matrix(arr)
mat.eliminate_zeros()
print(mat)
````

````python
mat = csr_matrix(arr)
mat.sum_duplicates()
print(mat)
````

````python
newarr = csr_matrix(arr).tocsc
print(newarr)
````

## Scipy Graphs
```scipy.sparse.csgraph``` module is used for graph data structure.

### Adjacency Matrix
Adjacency matrix is ```nxn``` matrix where ```n``` is the number of elements.  
and the values representing the connections between them.  
The Adjacency matrix would look like this:

````python
import numpy as np
from scipy.sparse.csgraph import connected_components
from scipy.sparse import csr_matrix

arr = np.array([
    [0, 1, 2],
    [1, 0, 0],
    [2, 0, 0]
])

newarr = csr_matrix(arr)
print(connected_components(newarr))
````

### Dijkstra 
Use the ```dijkstra``` method to find the shrotest path in graph from one element to another.  

It takes following arguments:  
1. **return_predecessors** - boolean (True to return the whole path of traversal otherwise False).
2. **indices** - index of the element to return all paths from that element only.
3. **limit** - max weight of the path.

````python
from scipy.sparse.csgraph import dijkstra
print(dijkstra(newarr, return_predecessors=True, indices=0))
````

### Floyd Warshall
Use the ```floyd_warshall()``` method to find the shortest path between all pairs of elements.  


````python
from scipy.sparse.csgraph import floyd_warshall
print(floyd_warshall(newarr, return_predecessors=True))
````

### Bellman Ford
The ```bellman_ford()``` method can also find the shortest path between all pairs of elements, but this method can handle negative weights as well.

````python
from scipy.sparse.csgraph import bellman_ford
print(bellman_ford(newarr, return_predecessors=True, indices=0))
````

### Depth First Order
The ```depth_first_order()``` method returns a depth first order traversal from a node.

The function take following arguments:
1. The graph
2. The starting element to traverse graph from.

````python
from scipy.sparse.csgraph import depth_first_order
arr = np.array([
    [0,1,0,1],
    [1,1,1,1],
    [2,1,1,0],
    [0,1,0,1]
])
newarr = csr_matrix(arr)
print(depth_first_order(arr,1))
````

### Breadth First Order
The ```breadth_first_order()``` method returns a depth first order traversal from a node.

The function take following arguments:
1. The graph
2. The starting element to traverse graph from.

````python
from scipy.sparse.csgraph import breadth_first_order
print(breadth_first_order(arr,1))
````

## Scipy spatial data
Spatial data refers to the data that can be represented in Geometric space.  
Scipy provides ```scipy.spatial``` module which has functions for working with spatial data.

### Triangulation
A Triangulation of a polygon is to divide the polygon into multiple triangles with which we can compute an area of the polygon.  
One method to generate these triangulations through points is the ```Delaunay()``` Triangulation.

````python
from scipy.spatial import Delaunay
import matplotlib.pyplot as plt

points = np.array([
    [2,4],
    [3,4],
    [3,0],
    [2,2],
    [4,1]
])
simplices = Delaunay(points).simplices
plt.triplot(points[:,0], points[:,1], simplices)
plt.scatter(points[:,0], points[:,1], color='r')
plt.show()
````

The ```simplices``` property creates a generalization of triangle notation.

### Convex Hull
A ```ConvexHull``` is the smallest polygon that covers all of the given points.

````python
from scipy.spatial import ConvexHull
hull = ConvexHull(points)
hull_points = hull.simplices

plt.scatter(points[:,0], points[:,1])
for simplex in hull_points:
    plt.plot(points[simplex,0],points[simplex,1],'k-')
plt.show()
````

### KDTrees
KDTrees are datastructure optimized for nearest neighbour queries.  
The ```KDTree()``` method returns KDTree object.  
The ```query()``` method returns the distance to the nearest neighbour and the location of neighbours.

````python
from scipy.spatial import KDTree
points = [(1,-1),(2,3),(-2,3),(2,-3)]
kdtree = KDTree(points)
res = kdtree.query((1,1))
print(res)
````

## Distance Matrix
There are many Distance Metrics used to find various types of distances between two points in data science, Euclidean distsance, cosine distsance etc.

The distance between two vectors may not only be the length of straight line between them, it can also be the angle between them from origin, or number of unit steps required etc.

Many of the Machine Learning algorithm's performance depends greatly on distance metrices. E.g. "K Nearest Neighbors", or "K Means" etc.

### Euclidean Distance


````python
from scipy.spatial.distance import euclidean
p1 = (1,0)
p2 = (10,2)
res = euclidean(p1, p2)
print(res)
````

### Cityblock Distance (Manhattan Distance)
It is the distance computed using 4 degrees of movement.

````python
from scipy.spatial.distance import cityblock
p1 = (1,0)
p2 = (10,2)
res = cityblock(p1, p2)
print(res)
````

### Cosine distance
It is the value of cosine angle between two points A and B

````python
from scipy.spatial.distance import cosine
p1 = (1,0)
p2 = (10,2)
res = cosine(p1, p2)
print(res)
````

### Hamming Distance
It is the proportion of bits where two bits are different.  
It is a way to measure distance for binary sequences.


````python
from scipy.spatial.distance import hamming
p1 = (True, False, True)
p2 = (False, True, True)
res = hamming(p1, p2)
print(res)
````

## Scipy Matlab Arrays
Scipy provides us with the module ```scipy.io```, which has functions for working with matlab arrays.

### Exporting data in Matlab format
The ```savemat()``` function is required to export data in Matlab format.  
The method takes following parameters:  
1. **filename** - the filename for saving data. 
2. **mdict** - a dictionary containing the data. 
3. **do_compression** - a boolean value that specifies whether to compress the result or not. Default False.

````python
from scipy import io
import numpy as np

arr = np.arange(10)
io.savemat('arr.mat',{"vec":arr})
````

### Import data from Matlab format
The ```loadmat``` function is used to import data from Matlab file.  
Parameter:  
1. **filename** - The filename of saved data.  
It will return the structured array where, keys are variable names and corresponding values are variable values.

````python
mydata = io.loadmat('arr.mat')
print(mydata)
````

````python
print(mydata['vec'])
````

## Scipy Interpolation
Interpolation is the method for generating points between two points.  
The method of filling values is called imputation.  
Scipy provides a module known as ```scipy.interpolate``` to deal with interpolation.

### 1D interpolation
The function ```interp1d()``` is used to interpolate a distribution with 1 variable.  
It takes ```x``` and ```y``` points and returns a callable function that can be called with new ```x``` and coressponding ```y```.

````python
from scipy.interpolate import interp1d
xs = np.arange(10)
ys = xs*2+1
interp_func = interp1d(xs,ys)
newarr = interp_func(np.arange(2.1, 3, 0.1))
print(newarr)
````

### SpLine interpolation
In 1D interpolation the points are fitted for *single curve* whereas in Spline interpolation the points are fitted against a *piecewise* function defined with polynomials called splines.  

The ```UnivariateSpline()``` function takes ```xs``` and ```ys``` and produce a callable function that can be called with new ```xs```.  

**Piecewise function** - A function that has different definition for different ranges.

````python
from scipy.interpolate import UnivariateSpline
xs = np.arange(10)
ys = xs**2 + np.sin(xs) + 1
interp_func = UnivariateSpline(xs, ys)
newarr = interp_func(np.arange(2.1,3,0.1))
print(newarr)
````

### Interpolation with Radial Basis Function
Radial basis function is a function that is defined to a corresponding to a fixed reference point.  
The ```Rbf()``` function takes ```xs``` and ```ys``` as arguments and produce a callable function that can be called with new ```xs```.

````python
from scipy.interpolate import Rbf
interp_func = Rbf(xs, ys)
newarr = interp_func(np.arange(2.1, 3, 0.1))
print(newarr)
````

## Scipy statistical significance tests
In statistics, statistical significance means that the result that was produced has a reason behind it, it was not reproduced randomly.  
Scipy provides us with a module called ```scipy.stats```.  


### Hypothesis in Statistics
Hypothesis is an assumption about a parameter in population.   

---

### Null Hypothesis
It assumes that the observation is not statistically significant.

---

### Alternate Hypothesis
It assumes that the observations are due to some reason.  
It alternates to Null Hypothesis.

---

### One Tailed Test
When our hypothesis is testing for one side of the value only.

---

### Two tailed test
When our hypothesis testing for both side of the values.

---

### Alpha value
Alpha value is the level of significance.

---

### P value
It tells how close to extremes the data actually is.

### T-tests (Two-Tailed Tests)
T-Tests is used to determine if there is a significant difference between means of two variables.  
The function ```ttest_ind()``` takes two samples of same size and produces a tuple of t-statistic and p-value.

````python
from scipy.stats import ttest_ind
v1 = np.random.normal(size=100)
v2 = np.random.normal(size=100)
res = ttest_ind(v1,v2)
print(res)
````

````python
res = ttest_ind(v1,v2).pvalue
print(res)
````

### KS-Test
KS Test is used to check if given values follows a distribution.  
The function takes the value to be tested, and CDF as two parameters.

---
A **CDF** can either be a string or a callable function that returns the probablity.  

---

It can be used for one tailed and tow tailed tests.  
By default it is two-tailed.

````python
from scipy.stats import kstest
v = np.random.normal(size=100)
res = kstest(v,'norm')
print(res)
````

### Statistical description of data
To see summary of values in array, we can use ```describe()``` function.  
It returns following description:
1. number of observations.
2. minimum and maximum values = minmax.
3. mean
4. variance
5. skewness
6. kurtosis

````python
from scipy.stats import describe
res = describe(v)
print(res)
````

### Normality Tests
Normality tests is based on Skewness and Kurtosis.  
The ```normaltest()``` function returns p value for null hypothesis.

---
### Skewness
A measure of symmetery of data.
1. 0 -> Normal distribution.
2. Negative -> Data is skewed left.
3. Postive -> Data is skewed right.

---

### Kutosis
A measure of whether the data is heavy or lightly tailed to normal distributions.
1. Positive -> Heavily tailed.
2. Negative -> Lightly tailed.

---

````python
from scipy.stats import skew, kurtosis
v=np.random.normal(size=100)
print(skew(v))
print(kurtosis(v))
````

````python
from scipy.stats import normaltest
print(normaltest(v))
````
