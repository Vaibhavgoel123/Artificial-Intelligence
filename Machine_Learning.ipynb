# Data types
We can split the data types into three major categories:  
1. **Numerical**: In this the data are numbers. They are further divided into two main categories:  
    - Discrete Data: Counted data that are limited to integers.
    - Continuous Data: Measured data that can be any number.
2. **Categorical**: In this the data are values that cannot be measured up against each other.
3. **Ordinal**: Similar to Categorical but in this data can be measured up against each other.

# Mean Median Mode
- **Mean** - The average value.
- **Median** - The mid point value.
- **Mode** - The most common value.

```python
import numpy as np
arr = np.array([1,2,3,4,5])
x = np.mean(arr)
print(x)
```

```python
y = np.median(arr)
print(y)
```

```python
from scipy import stats as sp
z=sp.mode(arr)
print(z)
```

## Variance
Variance is another factor that indicates how spread the values are.
